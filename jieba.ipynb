# 基本斷詞用法，使用預設詞庫

# 取出斷詞位置

# 取出文章中的關鍵詞

  -  jieba 使用了 tf-idf 方法來實作萃取出文章中關鍵詞的功能

```python
import jieba
import jieba.analyse

jieba.set_dictionary('C:\Users\BIG DATA\Anaconda2\Lib\site-packages\jieba\dict.txt.big.txt')

content = open('lyric.txt', 'rb').read()

#print "Input：", content

tags = jieba.analyse.extract_tags(content, 10)

print "Output："
print ",".join(tags)

#程式中的 jieba.analyse.extract_tags(content, 10)，
#就是告訴 jieba 我們要從這個文章中取出前 10 個 tf-idf 值最大的關鍵詞。
```

```python
import requests
from bs4 import BeautifulSoup as bs
import jieba
jieba.add_word('黃國昌')
dic = {}
index = 2089
for i in range(2050, 2089):
    res = requests.get('https://www.ptt.cc/bbs/Tech_Job/index{}.html'.format(i))
    soup = bs(res.text)
    for title in soup.select('.r-ent a'):
        for ele in jieba.cut(title.text):
            if not ele in dic:
                dic[ele]  = 1
            else:
                dic[ele] = dic[ele] + 1
            #print ele , dic[ele]
print dic
```

# 根據value值排序

```python
from operator import itemgetter
swd = sorted(dic.items(), key = itemgetter(1), reverse=True)
for ele in swd:
    print ele[0], ele[1]
```

# 應用--jieba斷詞後，依出現頻率排序
  
  1.  先在request把ptt的tech_job標題抓下 
  2.  用beautifulsoup將網頁標籤濾掉
  3.  用jieba斷文章裡的詞句，並存至dic中
  4.  用itemgetter對dic排序  (對標題作關鍵字排序)

```python
#import pyquery
import requests
from bs4 import BeautifulSoup as bs
import jieba
jieba.add_word('說一套')            #---動態新增字典，關掉後會消失
dic = {} 
url = 'https://www.ptt.cc/bbs/Tech_Job/index{}.html'
count = 0
for page in xrange(2138,2169):
    res = requests.get(url.format(page))
    soup = bs(res.text)
    for title in soup.select('.r-ent a'):
        count +=1
        for ele in jieba.cut(title.text):
            if not ele in dic:
                dic[ele]  = 1
            else:
                dic[ele] = dic[ele] + 1
            #print ele, dic[ele]
print '總文章數:'+str(count) 

from operator import itemgetter
        #dic可以用陣列表示，一個key 對一個value，
        #如下的意思是將dic裡面的每組資料當成陣列一個一個的取出
        #並以每個陣列的第二個元素[1]做排序-->key=itemgetter(1),
        #預設為生冪排序，所以reverse=True為降冪排序
swd = sorted(dic.items(), key = itemgetter(1), reverse=True)
for ele in swd:
    print ele[0], ele[1]
```
