# 基本斷詞用法，使用預設詞庫

```python
#-*- coding: utf-8 -*-
import jieba

sentence = "獨立音樂需要大家一起來推廣，歡迎加入我們的行列！"
print "Input：", sentence

words = jieba.cut(sentence, cut_all=False)
print "Output 精確模式 Full Mode："
for word in words:
    print word+'/',
```

# 取出斷詞位置

```python
import jieba

jieba.set_dictionary('C:\Users\BIG DATA\Anaconda2\Lib\site-packages\jieba\dict.txt.big.txt')

content = open('lyric.txt', 'rb').read()

#print "Input：", content

words = jieba.tokenize(unicode(content, 'utf-8'))

print "Output 精確模式 Full Mode："
for tk in words:
    print "word %s\t\t start: %d \t\t end:%d" % (tk[0],tk[1],tk[2])
```

# 取出文章中的關鍵詞

  -  jieba 使用了 tf-idf 方法來實作萃取出文章中關鍵詞的功能

```python
import jieba
import jieba.analyse

jieba.set_dictionary('C:\Users\BIG DATA\Anaconda2\Lib\site-packages\jieba\dict.txt.big.txt')

content = open('lyric.txt', 'rb').read()

#print "Input：", content

tags = jieba.analyse.extract_tags(content, 10)

print "Output："
print ",".join(tags)

#程式中的 jieba.analyse.extract_tags(content, 10)，
#就是告訴 jieba 我們要從這個文章中取出前 10 個 tf-idf 值最大的關鍵詞。
```

```python
import requests
from bs4 import BeautifulSoup as bs
import jieba
jieba.add_word('黃國昌')
dic = {}
index = 2089
for i in range(2050, 2089):
    res = requests.get('https://www.ptt.cc/bbs/Tech_Job/index{}.html'.format(i))
    soup = bs(res.text)
    for title in soup.select('.r-ent a'):
        for ele in jieba.cut(title.text):
            if not ele in dic:
                dic[ele]  = 1
            else:
                dic[ele] = dic[ele] + 1
            #print ele , dic[ele]
print dic
```

# 根據value值排序

```python
from operator import itemgetter
swd = sorted(dic.items(), key = itemgetter(1), reverse=True)
for ele in swd:
    print ele[0], ele[1]
```
